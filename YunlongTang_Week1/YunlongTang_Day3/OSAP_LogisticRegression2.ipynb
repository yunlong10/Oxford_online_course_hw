```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plotClass(X,y,p):
    plt.figure()
    for i in range(y.shape[1]):
        if y[0,i]==0:
            plt.plot(X[0,i],X[1,i],'r'+p)
        else:
            plt.plot(X[0,i],X[1,i],'b'+p)
            
    plt.show()
```

```python
#Q1
num_data=1000 # data points per class

X=np.random.uniform(-1,1,[2,num_data])
y=X[0,:]**2+X[1,:]**2-0.5>0
y=(y[None,:]).astype(np.int8)
plotClass(X,y,'o')
print(X.shape)
print(y.shape)
```

```python
num_basis=20
basis=[]

for i in range(num_basis):
    for j in range(num_basis):
        base=(X[0,:]**i)*(X[1,:]**j)
        basis.append(base)

basis=np.array(basis)

eps=1e-6

def get_z(X,W):
    return W@X

def sigmoid(z):
    return 1/(1+np.exp(-z))

def get_loss(y,yhat):
    return np.sum(-y*np.log(yhat+eps)-(1-y)*np.log(1-yhat+eps),axis=1)

w=np.random.random((1,400))

lr=0.001
ls=[]
for i in range(10000):
    z=get_z(basis,w)
    yhat=sigmoid(z)
    l=get_loss(y,yhat)

    ls.append(l)
    dz=yhat-y
    dw=dz@basis.T

    w=w-lr*dw

plotClass(X,yhat>=0.5,'x')
plt.show()
```
