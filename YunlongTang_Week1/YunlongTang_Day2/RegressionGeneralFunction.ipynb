```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x0=np.linspace(-1,1,100)

y0=np.sin(2*pi*5*x0)+0.0*np.random.randn(x0.size)
plt.plot(x0,y0)
```

```python
#f(x)=w_0 + w_1*x +w_2* x^2 +w_3*x^3 
reg=1e-11
x=x0[:,np.newaxis]
y=y0[:,np.newaxis]
num_basis=20
basis_pow=np.arange(num_basis+1)
basis_pow=basis_pow[np.newaxis,:]
A=x**basis_pow
w=np.linalg.inv(A.T@A+reg*np.eye(num_basis+1))@(A.T@y)
ynew=A@w
plt.plot(x,y,x,ynew)
```
