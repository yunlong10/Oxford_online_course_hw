```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plotClass(X,y,p):
    plt.figure()
    for i in range(y.shape[1]):
        if y[0,i]==0:
            plt.plot(X[0,i],X[1,i],'r'+p)
        else:
            plt.plot(X[0,i],X[1,i],'b'+p)
            
    plt.show()
```

```python
import torch
import torch.nn as nn
from torch import optim
from torch.optim.lr_scheduler import StepLR
class Net(nn.Module):
    def __init__(self,ni,nh,no):
        super(Net,self).__init__()
        self.l1=nn.Linear(ni,nh)
        self.l2=nn.Linear(nh,no)

    def forward(self,x):
        x=torch.sigmoid(self.l1(x))
        x=torch.sigmoid(self.l2(x))
        return x
        
class Net2(nn.Module):
    def __init__(self,ni,nh,no):
        super(Net2,self).__init__()
        self.l1=nn.Linear(ni,nh)
        self.l2=nn.Linear(nh,nh)
        self.l3=nn.Linear(nh,no)
        

    def forward(self,x):
        x=torch.relu(self.l1(x))
        x=torch.relu(self.l2(x))
        x=torch.sigmoid(self.l3(x))
        return x
```

```python
#Q1
num_data=1000 # data points per class

X=np.random.uniform(-1,1,[2,num_data])
y=X[0,:]**2+X[1,:]**2-0.5>0
y=(y[None,:]).astype(np.int8)
plotClass(X,y,'o')
print(X.shape)
print(y.shape)
```

```python
ni,nh,no=2,10,1
lr=0.1
model=Net2(ni,nh,no)
ls=[]
loss_fn=nn.BCELoss()
optimizer=optim.SGD(model.parameters(),lr=lr,momentum=0.5)
scheduler=StepLR(optimizer,step_size=2000,gamma=0.8)
num_epoches=10000

X_=torch.Tensor(X.T)
y_=torch.Tensor(y.T)

for i in range(num_epoches):
    optimizer.zero_grad()
    yhat=model(X_)
    loss=loss_fn(yhat,y_)

    loss.backward()
    optimizer.step()
    scheduler.step()
    ls.append(loss.item())

plt.plot(ls)
plotClass(X,yhat.detach().numpy().T>0.5,'x')
```
